```python
import numpy as np
input_data = np.array([[12,-19,9],
                      [2.7,0.7,19.9],
                      [1,0,39]])
print(input_data)
```

```python
import matplotlib.pyplot as plt
plt.plot(input_data)
plt.show()
```

```python
from sklearn import preprocessing
binary_data=preprocessing.Binarizer(threshold=12).transform(input_data)
print(binary_data)
```

```python
import matplotlib.pyplot as plt
plt.plot(binary_data)
plt.show()
```

```python
Minmax_data=preprocessing.MinMaxScaler(feature_range=(10,20)).fit_transform(input_data)
print(Minmax_data)
```

```python
normalized_data= preprocessing.normalize(input_data,norm='l1')
print(normalized_data)
```

```python
normalized_data= preprocessing.normalize(input_data,norm='l2')
print(normalized_data)
```
